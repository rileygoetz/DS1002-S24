```python
import matplotlib.pyplot as plt

def red_filter(user_image):
  red_filter_only = user_image.copy()
  red_filter_only[:,:,1:] = 0
  return red_filter_only

def green_filter(user_image):
  green_filter_only = user_image.copy()
  green_filter_only[:,:,[0,2]] = 0
  return green_filter_only

def blue_filter(user_image):
  blue_filter_only = user_image.copy()
  blue_filter_only[:,:,:2] = 0
  return blue_filter_only

def filter_out_blue(user_image):
  no_blue = user_image.copy()
  no_blue[:,:,2] = 0
  return no_blue

def filter_out_green(user_image):
  no_green = user_image.copy()
  no_green[:,:,1] = 0
  return no_green

def filter_out_red(user_image):
  no_red = user_image.copy()
  no_red[:,:,0] = 0
  return no_red

url = input("https://flowermag.com/wp-content/uploads/2017/11/magnolia.jpg")
url = 'https://flowermag.com/wp-content/uploads/2017/11/magnolia.jpg'
response = requests.get(url)

image = Image.open(BytesIO(response.content))
user_image = np.array(image)

red_filtered_image = red_filter(user_image)
red_image = Image.fromarray(red_filtered_image, 'RGB')

green_filtered_image = green_filter(user_image)
green_image = Image.fromarray(green_filtered_image, 'RGB')

blue_filtered_image = blue_filter(user_image)
blue_image = Image.fromarray(blue_filtered_image, 'RGB')

no_blue_image = filter_out_blue(user_image)
RG_image = Image.fromarray(no_blue_image, 'RGB')

no_green_image = filter_out_green(user_image)
RB_image = Image.fromarray(no_green_image, 'RGB')

no_red_image = filter_out_red(user_image)
GB_image = Image.fromarray(no_red_image, 'RGB')

fig,axes = plt.subplots(2,4, figsize=(18,10))
axes[0,0].imshow(user_image)
axes[0,0].set_title('Original')

axes[0,1].imshow(red_image)
axes[0,1].set_title("Red Filter")

axes[0,2].imshow(green_image)
axes[0,2].set_title("Green Filter")

axes[0,3].imshow(blue_image)
axes[0,3].set_title("Blue Filter")

axes[1,0].imshow(RG_image)
axes[1,0].set_title("No Blue Image")

axes[1,1].imshow(RB_image)
axes[1,1].set_title("No Green Image")

axes[1,2].imshow(GB_image)
axes[1,2].set_title("No Red Image")

for ax in axes.ravel():
  ax.axis('off')

plt.show()

combined_filters = red_filtered_image + green_filtered_image + blue_filtered_image
are_the_same = np.array_equal(user_image, combined_filters)
print("The Original Image is Identical to the Combined Filtered Images;", are_the_same)
```
